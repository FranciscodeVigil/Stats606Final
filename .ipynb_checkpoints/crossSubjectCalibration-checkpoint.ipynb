```python
import numpy as np
from sklearn.manifold import MDS
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import NearestNeighbors
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Data = pd.read_csv("P300CrossTesting_K_Protocol(Sheet1).csv")
Data = Data.to_numpy()
Acc_Matrix = Data[0:,5:]
dims = Acc_Matrix.shape
Acc_Matrix = Acc_Matrix.astype(float)

# Transforming the accuracy matrix into something that better invokes distance
ep = 0.00001
Distance_Matrix = (1 - (Acc_Matrix / 100.0)) * 100

Symetrized_Mat = (Distance_Matrix + Distance_Matrix.T)/2

# The calibration accuracy is rarely 100 for the subject the weights were generated from
# but to make the matrix appear like it was generated from a metric we make the diagnol 0 

np.fill_diagonal(Symetrized_Mat, 0)

# Some attempts to see if rescaling would yield better results
preProcessScale = MinMaxScaler()
ScaledDist = preProcessScale.fit_transform(Symetrized_Mat)
```

### Attempt to find an eps for DBSCAN
The following is the K-distance graph for the distance matrix to see if there is a ideal eps value

```python
neigh = NearestNeighbors(n_neighbors=2, metric='precomputed')
neighFit = neigh.fit(ScaledDist)
distances, indices = neighFit.kneighbors() 
distances = np.sort(distances[:, 1])
plt.figure(figsize=(8,5))
plt.plot(distances)
plt.title('K-distance Graph (DBSCAN tuning)')
plt.ylabel('Distance to nearest neighbor')
plt.xlabel('Points sorted by distance')
plt.grid(True)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.6, min_samples=5,metric = 'precomputed')
dbscan.fit(ScaledDist)
labels = dbscan.labels_

print("Cluster assignments:", labels)
```

```python
dists = Symetrized_Mat[np.triu_indices_from(Symetrized_Mat, k=1)]

plt.hist(dists, bins=30, edgecolor='black')
plt.title("Distribution of Pairwise Distances")
plt.xlabel("Distance")
plt.ylabel("Count")
plt.show()
```

### This is a hierarchical clustering model useing the Average distance method

```python
condensed = squareform(Symetrized_Mat)
linksAv = linkage(condensed, method ='average')

plt.figure(figsize=(10,5))

# Generating the dendrogram to look for optimal values to split the groups
dendrogram(linksAv, labels=[f"S{i}" for i in range(Symetrized_Mat.shape[0])])
plt.title("Hierarchical Clustering Dendrogram (Method: Average)")
plt.xlabel("Subjects")
plt.ylabel("Distance")
plt.show()
```

```python
labelsAv = fcluster(linksAv, t = 80, criterion='distance')
groupsAv = list(set(labelsAv))
print("Unique Groups:",groupsAv)
print("Cluster labels:", labelsAv)
```

### This instead uses the complete metric in place of the average distance metric

```python
linksCmp = linkage(condensed, method ='complete')
plt.figure(figsize=(10,5))

dendrogram(linksCmp, labels=[f"S{i}" for i in range(Symetrized_Mat.shape[0])])
plt.title("Hierarchical Clustering Dendrogram (Method:Complete)")
plt.xlabel("Subjects")
plt.ylabel("Distance")
plt.show()
```

```python
labelsCmp = fcluster(linksCmp, t = 87, criterion='distance')
groupsCmp = list(set(labelsCmp))
print("Unique Groups:",groupsCmp)
print("Cluster labels:", labelsCmp)
```

```python
nSub = Symetrized_Mat.shape[0]
predictedClust = []

# This is a smaller code chunk that allows us to do leave-one-out accuracy checking for a fixed t value that was chosen
# from the information in the dendrogram
for thisSub in range(nSub):
    mask = np.arange(nSub) != thisSub
    trainMat = Symetrized_Mat[np.ix_(mask, mask)]
    
    linkageTrain = linkage(squareform(trainMat), method='complete')
    clusterTrain = fcluster(linkageTrain, t=87, criterion='distance')

    testDist = Symetrized_Mat[thisSub, mask] 
    
    clusterMeans = []
    for cluster in np.unique(clusterTrain):
        members = np.where(clusterTrain == cluster)[0]
        meanDist = np.max(testDist[members])
        clusterMeans.append((cluster, meanDist))
    
    predictedClust.append(min(clusterMeans, key=lambda x: x[1])[0])
    
    print(f"Subject {thisSub} would be assigned to cluster {predictedClust[thisSub]}")
```

```python
sum(labelsCmp == predictedClust) / 39
```

```python
adjusted_rand_score(labelsCmp,predictedClust)
```

```python
# This does the same thing as the privious large code block but instead looks at the leave one out accuracy
# for all the integer values of t 0:100 then calculates the adjusted random score to see the cluster assigment accuracy

adjRand = []
silhouette = []
for cutSpot in np.arange(1, 101):
    labelsCmpMany = fcluster(linksCmp, t = cutSpot, criterion='distance')
    if len(set(labelsCmp)) < 2:
        print(f"Warning: only {len(set(labelsCmp))} cluster(s) at t = {cutSpot}")

    if 1 < len(set(labelsCmpMany)) < Symetrized_Mat.shape[0]:
        silhouette.append(silhouette_score(Symetrized_Mat, labelsCmpMany, metric="precomputed"))
    else:
        print(f"Skipping silhouette: {len(set(labelsCmpMany))} cluster(s) at t = {cutSpot}")
        silhouette.append(np.nan) 
    
    nSub = Symetrized_Mat.shape[0]
    predictedClust = []

    for thisSub in range(nSub):
        #Creates a mask which includes all but the chosen point.
        mask = np.arange(nSub) != thisSub
        trainMat = Symetrized_Mat[np.ix_(mask, mask)]
    
        linkageTrain = linkage(squareform(trainMat), method='complete')
        clusterTrain = fcluster(linkageTrain, t=cutSpot, criterion='distance')
    
        testDist = Symetrized_Mat[thisSub, mask] 
    
        clusterMeans = []
        for cluster in np.unique(clusterTrain):
            members = np.where(clusterTrain == cluster)[0]
            meanDist = np.mean(testDist[members])
            clusterMeans.append((cluster, meanDist))
    
        predictedClust.append(min(clusterMeans, key=lambda x: x[1])[0])
    
    ari = adjusted_rand_score(labelsCmpMany,predictedClust)
    adjRand.append(ari)
```

```python
plt.plot(silhouette)
plt.title("t vs Silhouette Score")
plt.xlabel("t value")
plt.ylabel("Silhouette Score")
```

```python
#Gets rid of the degenerate cases where there is only 1 cluster
trimmedAdjRand = adjRand[:-3]
print(max(trimmedAdjRand))
trimmedAdjRand.index(max(trimmedAdjRand))
```

```python
# Visualizes the optimal t values for leave one out prediction accuracy
plt.plot(np.arange(1, 98),trimmedAdjRand)
plt.title("t vs Adjusted Rand Score")
plt.xlabel("t value")
plt.ylabel("Calculated Adjusted Random Score")
```

```python
#Creates a MDS embedding so Kmeans can be properly used
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42)
embeddedPoints = mds.fit_transform(Symetrized_Mat)
```

```python
print(embeddedPoints)
```

```python
#Analysis of the inertial values in order to find an optimal number of clusters

inertiaValues = []

for k in range(1, 11): 
    kmeans = KMeans(n_clusters=k, random_state=67)
    kmeans.fit(embeddedPoints)
    inertiaValues.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertiaValues, marker='o')
plt.title('Search for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, n_init='auto',random_state=67)
kcloud = kmeans.fit_predict(embeddedPoints)

print(kcloud)

plt.scatter(Symetrized_Mat[:, 0], Symetrized_Mat[:, 1], c=kcloud, cmap='viridis', edgecolors='k')
plt.title("KMeans Clustering")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
#LOOCV for the clusters
leftResults = []
for pointNum in range(kcloud.shape[0]):
    
    
    cloudTrain = np.delete(embeddedPoints, pointNum, axis=0)
    
    kmeans.fit_predict(cloudTrain)

    centers = kmeans.cluster_centers_

    testPoint = embeddedPoints[pointNum].reshape(1, -1)
    closestCenterNum = pairwise_distances_argmin_min(testPoint, centers)[0][0]
        
    leftResults.append(closestCenterNum)
```

```python
print(leftResults)
```

```python
kMeansAcc = sum(leftResults == kcloud) / kcloud.shape[0]
print(kMeansAcc)
adjusted_rand_score(leftResults,kcloud)
```
